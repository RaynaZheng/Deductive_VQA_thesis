```python
import sys 
import os 
import json

with open('/Users/raynazheng/Dropbox/Deductive visual question answering with spatial ontology/data/train_sceneGraphs.json') as file:
    data = json.load(file) 
```

```python
image_id_list = [] #the length of this list should be 126
```

```python
for i_id,info in data.items():
    objects = info['objects']
    for o,o_info in objects.items():
        relations = o_info['relations'] #relations is a list of dictionaries
        for re in relations:
            if re['name'] == 'next to':
                image_id_list.append(i_id)
            else:
                continue
image_id_list
```

```python
relation_dict = {}
for s,i in data.items():
    object_info = i['objects']
    obj_name = {}
    for obj_id, name in object_info.items():
        obj_name[obj_id] = name['name']
    for idx, info in object_info.items():
        rela = info['relations']
        sub = info['name']
        if rela:
            for r in rela:
                name = rela[0]['name']
                obj = rela[0]['object']
                if name == 'next to':
                    if s in relation_dict:
                        object_name = obj_name[obj] if obj in obj_name else None
                        tmp = {'subjects': sub, 'objects': object_name}
                        relation_dict[s].append(tmp)
                    else:
                        tmp = {'subjects': sub, 'objects': obj}
                        relation_dict[s] = [tmp]

with open('nextto_in_GQA.txt','w') as text: 
      text.write(str(relation_dict))
```

```python
for image in image_id_list:
    info = data[image]['objects']
    # create a dictionary for all the objects in the image
    # extract the name for each object
    obj_dict = {}
    for o, obj in info.items():
        obj_dict[o] = obj['name']
    
    # open file in write mode
    file_name = f'{image}.txt'
    with open(r'/Users/raynazheng/Dropbox/Deductive visual question answering with spatial ontology/data/nextto_samples/'+file_name, 'w') as fp:
        # write each relationship sub+predicate+obj
        for o, obj in info.items():
            subject = obj_dict[o]
            for re in obj['relations']:
                predicate = re['name']
                object_name = obj_dict[re['object']]
                sentence = subject+' '+predicate+' '+object_name
                fp.write("%s\n" % sentence)
```
